```python
import os
import requests
import jwt  # pip install PyJWT

JWT_SECRET = os.getenv("JWT_SECRET", "change-me")
JWT_ALGORITHM = "HS256"
SOURCE_URL = "https://example.com/your-endpoint"  # TODO: replace

def create_jwt_with_et_from_url() -> str:
    # 1. Call the URL
    resp = requests.get(SOURCE_URL, timeout=5)
    resp.raise_for_status()

    # 2. Parse JSON and extract `et`
    data = resp.json()
    et = data.get("et")
    if et is None:
        raise ValueError("Response JSON has no 'et' field")

    # 3. Build JWT payload
    payload = {
        "et": et,
        # add other claims if needed
        # "sub": "user-id",
    }

    # 4. Sign and return token
    token = jwt.encode(payload, JWT_SECRET, algorithm=JWT_ALGORITHM)
    return token
```
